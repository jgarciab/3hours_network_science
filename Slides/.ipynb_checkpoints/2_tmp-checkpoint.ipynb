```python
#Netowrks
import networkx as nx
import igraph as ig

#Viz
import pylab as plt
import seaborn as sns


import pandas
import numpy as np

#%matplotlib notebook
%matplotlib inline
```

```python


# data on florentine marriage families in the XV century
G = nx.florentine_families_graph()

plt.figure(figsize=(8,4))
# Set up a seed so we have the same graph
np.random.seed(3) 

g = ng.InteractiveGraph(G, node_layout = "spring", #arrows=True, 
                     node_size=2, node_color = "cornflowerblue", 
                     node_labels=True,  node_shape = "d", node_edge_width = 0,
                     node_label_offset=0.05, node_label_fontdict=dict(size=8),
                     edge_width=1, edge_color = "gray")
```

```python
degree = np.array([deg for (node, deg) in G.degree()])
```

```python
A = nx.to_scipy_sparse_array(G, nodelist=list(G.nodes()), weight=1)
N = len(G.nodes())

# show the adjacency matrix
plt.spy(A)

# add the right 
plt.xticks(range(N), G.nodes(), rotation=90)
plt.yticks(range(N), G.nodes())

# make sure the figure fits 
plt.tight_layout()
```

```python
# Exercise 1: Sum of neighbors degree
sum_degree = A @ degree

# add labels for printing
list(zip(G.nodes(), sum_degree))

```

```python
A.sum(1)
```

```python
# Exercise 2: Average neighbors degree
avg_degree = A @ degree / np.sum(A, 1)

list(zip(G.nodes(), avg_degree))
```

```python
# Exercise 3: Number of nodes at length <2
path_2 = A**2
path_2.setdiag(0)
path_2 += A # Add direct edges
path_2 = path_2>0


list(zip(G.nodes(), path_2))


g = ng.InteractiveGraph(G, node_layout = "spring", #arrows=True, 
                     node_size=2, node_color = "cornflowerblue", 
                     node_labels=True,  node_shape = "d", node_edge_width = 0,
                     node_label_offset=0.05, node_label_fontdict=dict(size=8),
                     edge_width=1, edge_color = "gray")
plt.show()

# show the adjacency matrix
plt.spy(path_2)

# add the right 
plt.xticks(range(N), G.nodes(), rotation=90)
plt.yticks(range(N), G.nodes())

# make sure the figure fits 
plt.tight_layout()
plt.show()

path_2 = path_2.sum(1)
```

```python
# Exercise 3: Number of nodes at length <3
path_3 = A**3 + path_2
#path_2.setdiag(0) #no needed, you can't go back in 3 steps

path_3 = (path_3 > 0).sum(1)

list(zip(G.nodes(), path_3))
```

```python
# Exercise 4: Eigenvector centrality, manually

weight = np.ones(N)



for i in range(100):
    # calculate power
    weight = A @ weight 

    # normalize uwing the geometrical mean
    weight = weight / np.sqrt(np.sum(weight**2))
    
list(zip(G.nodes(), weight))

#compare with nx.eigenvector_centrality(G)
```

```python
# Exercise 5: Number of triangles
path_3 = A**3
#path_2.setdiag(0) #no needed, you can't go back in 3 steps

path_3 = path_3.diagonal() / 2

# number of friedns who are friends among themselves
combinations = (A.sum(1) * (A.sum(1)-1))/2
local_cl = path_3 / combinations
local_cl[~np.isfinite(local_cl)] = 1

np.random.seed(3)

g = ng.InteractiveGraph(G, node_layout = "spring", #arrows=True, 
                     node_size=2, node_color = "cornflowerblue", 
                     node_labels=True,  node_shape = "d", node_edge_width = 0,
                     node_label_offset=0.05, node_label_fontdict=dict(size=8),
                     edge_width=1, edge_color = "gray")
plt.show()

print(list(zip(G.nodes(), path_3)))
print(list(zip(G.nodes(), local_cl)))
```

```python
# Exercise 5: Local assortativity (e.g. degree)
```

```python
degree = A.sum(1)
diff_degree = (A * degree).todense().T - degree
diff_degree[(A==0).todense()] = 0
assort = np.abs(diff_degree).sum(1)/degree

assort = assort/np.sum(assort)


assort = (1+nx.degree_assortativity_coefficient(G))/N - assort



list(zip(G.nodes(), np.round(assort,2)))
```
